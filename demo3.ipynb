```python
from abc import ABC, abstractmethod

class SinhVien(ABC):
    uni = "Trường đh công nghiệp tphcm"
    def __init__(self, mssv, hoten):
        self.mssv = mssv
        self.hoten = hoten
        self.__cccd = '123'
    
    def get_cccd(self):
        return self.__cccd
    
    def set_cccd(self, cccd):
        self.__cccd = cccd

    def InthongTin(self):
        print(f"{self.mssv} - {self.hoten} - {self.uni}")
    
    @abstractmethod
    def TinhDiemTrungBinh(self):
        pass
```

```python
class KSTN(SinhVien):
    def __init__(self, mssv, hoten, diemCTRR, diemTA):
        super().__init__(mssv, hoten)
        self.diemCTRR = diemCTRR
        self.diemTA = diemTA

    def InthongTin(self):
        super().InthongTin()
        print(f"Điểm CTRR {self.diemCTRR} - Điểm TA {self.diemTA}")

    def TinhDiemTrungBinh(self):
        dtb = (self.diemCTRR + self.diemTA) / 2
        print(f"Điểm trung bình: {dtb}")
```

```python
class TCTA(SinhVien):
    def __init__(self, mssv, hoten, diemAV3, diemAV4):
        super().__init__(mssv, hoten)
        self.diemAV3 = diemAV3
        self.diemAV4 = diemAV4

    def InthongTin(self):
        super().InthongTin()
        print(f"Điểm AV3 {self.diemAV3} - Điểm AV4 {self.diemAV4}")

    def TinhDiemTrungBinh(self):
        dtb = (self.diemAV3 + self.diemAV4) / 2
        print(f"Điểm trung bình: {dtb}")
```

```python
svTCTA = TCTA('SV01', 'Hung', 7, 9)
```

```python
svTCTA.InthongTin()
svTCTA.TinhDiemTrungBinh()
```
