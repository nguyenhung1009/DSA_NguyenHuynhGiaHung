```python
from abc import ABC, abstractmethod

class Person(ABC):
    def __init__(self, id, name):
        self._id = id
        self._name = name

    def get_id(self):
        return self._id

    def get_name(self):
        return self._name

    def set_name(self, name):
        self._name = name

    @abstractmethod
    def get_role(self):
        pass
    @abstractmethod
    def to_dict(self):
        pass
```

```python
class Student(Person):
    def __init__(self, id, name, gpa):
        super().__init__(id, name)
        self.__gpa = gpa 

    def set_gpa(self, gpa_new):
        self.__gpa = gpa_new

    def get_gpa(self):
        return self.__gpa
    
    def get_role(self):
        return "Student"
    
    def to_dict(self):
        return {
            "id": self.get_id(),
            "name": self.get_name(),
            "gpa": self.get_gpa()
        }
    
```

```python
class UnderStudent(Student):
    def __init__(self, id, name, gpa, year):
        super().__init__(id, name, gpa)
        self.__year = year

    def get_role(self):
        return "UnderStudent"
    
    def to_dict(self):
        pass
    
```

```python
class GraduatesStudent(Student):
    def __init__(self, id, name, gpa, thesis):
        super().__init__(id, name, gpa)
        self.__thesis = thesis
    
    def get_role(self):
        return "GraduatesStudent"
    
    def to_dict(self):
        pass
```

```python
class Manager:
    def __init__(self):
        self._lstStud = []

#thêm
    def add_Student(self, s):
        self._lstStud.append(s)

#tìm
    def find_Student(self, id):
        for s in self._lstStud:
            if s.get_id() == id:
                return s
        return None
    
#xoá
    def del_Student(self, id):
        s = self.find_Student(id)
        if s:
            self._lstStud.remove(s)
            print(f"Đã xoá sinh viên {id}")
        else:
            print("Không tìm thấy sinh viên")

#cập nhật
    def update_Student(self, id, name, gpa):
        s = self.find_Student(id)
        if s:
            if name:
                s.set_name(name)
            if gpa:
                s.set_gpa(gpa)
            print(f"Đã cập nhật sinh viên {id}")
        else:
            print("Không tìm thấy sinh viên")
    


#in thông in
    def print_student(self):
        if not self._lstStud:
            print("Danh sách rỗng")
            return 
        ##for s in self._lstStud:
        ##print(s.to_dict(), "Role:", s.get_role())
        print("Danh sách sinh viên")
        print("----")
        for s in self._lstStud:
            print(f"Mã sinh viên: {s.get_id()}")
            print(f"Tên sinh viên: {s.get_name()}")
            print(f"GPA: {s.get_gpa()}")
            print(f"Role: {s.get_role()}")
            print("----")


    ## thêm , xoá , cập nhật, in thông tin 
    ## làm menu cho phép tự chọn chức năng
            
```

```python
def menu():
    manager = Manager()

    while True:
        print("\n===== MENU QUẢN LÝ SINH VIÊN =====")
        print("1. Thêm sinh viên")
        print("2. Tìm sinh viên")
        print("3. Xoá sinh viên")
        print("4. Cập nhật sinh viên")
        print("5. In danh sách sinh viên")
        print("0. Thoát")

        choice = input("Chọn chức năng: ")

        if choice == "1":
            id = input("Nhập ID: ")
            name = input("Nhập tên: ")
            gpa = float(input("Nhập GPA: "))
            s = Student(id, name, gpa)
            manager.add_Student(s)

        elif choice == "2":
            id = input("Nhập ID cần tìm: ")
            s = manager.find_Student(id)
            if s:
                print(s.to_dict())
            else:
                print("Không tìm thấy")

        elif choice == "3":
            id = input("Nhập ID cần xoá: ")
            manager.del_Student(id)

        elif choice == "4":
            id = input("Nhập ID cần cập nhật: ")
            name = input("Tên mới (Enter bỏ qua): ")
            gpa = input("GPA mới (Enter bỏ qua): ")

            if gpa != "":
                gpa = float(gpa)
            else:
                gpa = None

            manager.update_Student(id, name if name != "" else None, gpa)

        elif choice == "5":
            manager.print_student()

        elif choice == "0":
            print("Thoát chương trình")
            break

        else:
            print("Chọn sai!")
```
