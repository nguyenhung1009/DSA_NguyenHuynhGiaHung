```python
products = [{
    "id": 1,
    "name": "iphone",
    "price": 100,
    "description": "iphone best"
},
{
    "id": 2,
    "name": "ipad",
    "price": 120,
    "description": "ipad best"
}]
print(products)
```

```python
def Xuat_Danh_Sach():
    print("Danh sách sản phẩm")
    print("----")
    for p in products:
        print(f"Mã sản phẩm: {p['id']}")
        print(f"Tên sản phẩm: {p['name']}")
        print(f"Giá sản phẩm: {p['price']}")
        print(f"Mô tả sản phẩm: {p['description']}")
        print("----\n")

Xuat_Danh_Sach()
```

```python
# Thêm 1 sản phẩm
def them_sp():
    id = products[-1]["id"] + 1 if products else 1
    ten = input("Nhập tên sản phẩm: ")
    gia = int(input("Nhập giá sản phẩm: "))
    mo_ta = input("Nhập mô tả sản phẩm: ")

    new_sp = {
           "id": id,
            "name": ten,
            "price": gia,
            "description": mo_ta
}

    products.append(new_sp)    
them_sp()
     
```

```python
# Thêm nhiều sản phẩm
def them_nhieu_sp():
    n = int(input("Nhập số lượng sản phẩm: "))
    for i in range(n):
        id = products[-1]["id"] + 1 if products else 1
        ten = input("Nhập tên sản phẩm: ")
        gia = int(input("Nhập giá sản phẩm: "))
        mo_ta = input("Nhập mô tả sản phẩm: ")

        new_sp = {
                "id": id,
                "name": ten,
                "price": gia,
                "description": mo_ta
    }

        products.append(new_sp)  
them_nhieu_sp()
```

```python
# Xoá sản phẩm
def xoa():
    find_id = int(input("Nhập id muốn xoá: "))

    found = False

    for p in products:
        if p['id'] == find_id:
            products.remove(p)
            print("Đã xoá thành công")
            found = True
            break
    if not found:
        print(f"Không có {id} đó để xoá")

xoa()
```

```python
# Tìm kiếm sản phẩm theo tên 
def tim_kiem():
    find_ten = input("Nhập tên muốn tìm: ").lower()

    found = False

    for p in products:
        if find_ten in p['name'].lower():
            print("Sản phẩm tìm được: ")
            print(p)
            found = True
        
    if not found:
        print(f"Không tìm thấy sp có {find_ten}")
tim_kiem()
```

```python
# Cập nhật sản phẩm
def capnhat():
    found = False
    id = int(input("Nhập id muốn cập nhật: "))
    for p in products:
        if p['id'] == id:
            name = input("Nhập tên mới: ")
            price = input("Nhập giá mới: ")
            description = input("Nhập mô tả mới: ")

            if name:
                p['name'] = name
            if name:
                p['price'] = price
            if description:
                p['description'] = description

            print(f"Cập nhật thành công mã {id}")
            found = True
    if not found:
        print(f"Không tìm thấy mã {id}")
```
