```python
def KiemTraSNT():
    if(n < 2):
        return False
    for i in range(2, n):
        if(n % i == 0):
            return False
    return True

n = int(input("Nhập số nguyên: "))
if KiemTraSNT():
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không phải số nguyên tố")   
```
