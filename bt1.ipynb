```python
class SinhVien():
    def __init__(self, id, name, age = None, gpa = None):
        self._id = id
        self._name = name
        self._age = age
        self._gpa = gpa

    def InThongTin(self):
        return (f"id:{self._id}, name:{self._name}, age:{self._age}, gpa:{self._gpa}")
```

```python
class SinhVienChinhQuy(SinhVien):
    def __init__(self, id, name, age = None, gpa = None, noi_dao_tao = None):
        super().__init__(id, name, age, gpa)
        self.noi_dao_tao = noi_dao_tao

    def InThongTin(self):
        return f"{super().InThongTin()} - nơi đào tạo: {self.noi_dao_tao}"
```

```python
class SinhVienTaiChuc(SinhVien):
    def __init__(self, id, name, age = None, gpa = None, noi_dao_tao = None, chuyen_nganh =None):
        super().__init__(id, name, age, gpa, noi_dao_tao)
        self.chuyen_nganh = chuyen_nganh

    def InThongTin(self):
        return f"{super().InThongTin()} - chuyên ngành: {self.chuyen_nganh}"
```

```python
from abc import ABC , abstractmethod

class BaseSV(ABC):
    
    @abstractmethod
    def is_empty(self):
        pass

    @abstractmethod    
    def push(self, sv):
        pass

    @abstractmethod
    def pop(self):
        pass
```

```python
class StackSV(BaseSV):
    def __init__(self):
        self.lst_stack = []

    def is_empty(self):
        return len(self.lst_stack) == 0
    
    def push(self, sv):
        self.lst_stack.append(sv)

    def pop(self):
        ##if self.lst_stack.is_empty():
        ##   return None
        ##return self.lst_stack.pop()

        return None if self.is_empty() else self.lst_stack.pop()
```

```python
class Manage:
    def __init__(self):
        self.ds = StackSV()

    def add_student(self, sv):
        self.ds.push(sv)

    def display_student(self):
        if self.ds.is_empty():
            return "Danh sách rỗng"
        
        return "\n".join((sv.InThongTin()) for sv in self.ds.lst_stack)
    
    def delete_all(self):
        self.ds.lst_stack.clear()
        return "delete all"

    def delete(self, id):
        for sv in self.ds.lst_stack:
            self.ds.lst_stack.remove(sv)
            if sv._id == id:
                return "Đã xoá"
        return "Không tìm thấy sinh viên"
    
    def findSV(self, id):
        for sv in self.ds:
            if sv.id == id:
                return sv
            return None
        
    def updateSV(self, id, newSV):
        for sv in self.ds:
            self.name = newSV.name 
            self._age = newSV.age
            self._gpa = newSV.gpa
            return True
        return False


```
